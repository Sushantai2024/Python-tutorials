## Pandas Tutorial 

Pandas is an open-source data maniplation and analysis library for Python.It provides data structures like series (1D labeled array) and Dataframe (2D labeled data structure), which are designed for structure data, such as tables.

 It is widely used for tasks such as data cleaning, preparation, exploration, and analysis.


### Here are the two key structures in Pandas:

1.	__Series__: A one-dimensional labeled array capable of holding any data type (integers, strings, floating-point numbers, etc.). It is similar to a column in a table or a single list in Python.


2.	__DataFrame__: A two-dimensional labeled data structure, similar to a table in a database or a spreadsheet, with rows and columns. Each column in a DataFrame is a Series.


#### Example:

```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {'Name': ['John', 'Jane', 'Tom'],
        'Age': [23, 24, 22],
        'City': ['New York', 'Paris', 'London']}

df = pd.DataFrame(data)
print(df)
```
